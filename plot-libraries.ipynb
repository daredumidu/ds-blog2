```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = ["date_establishment", "date_destruction", "cultural_loss", "human_headline", "human_stories", "obituary", "caption"],
      color = "green"
    ),
   link = dict(
      source = [0, 1, 0, 2, 3, 3], # indices correspond to labels, eg A1, A2, A1, B1, ...
      target = [2, 3, 3, 4, 4, 5],
      value = [8, 4, 2, 8, 4, 2]
  ))])
fig.update_layout(title_text="Basic Sankey Diagram", font_size=10)
fig.show()
```

```python
df = df.dropna(subset=['numbered_loss'])
```

```python
# Plot histogram

fig = px.histogram(df, x="numbered_loss", nbins=20, title="Mapping Loss — Libraries and Lives in Palestine")
fig.update_layout(xaxis_title="Human Loss", yaxis_title="count")
fig.show()
```

```python
fig = px.histogram(df, x="date_establishment")
fig.update_layout(bargap=0.2)
fig.show()
```

```python
fig = px.histogram(df, x="date_destruction", category_orders=dict(day=["Thur", "Fri", "Sat", "Sun"]))
fig.show()
```

```python
fig = px.histogram(df, x="obituary", color="gender")
fig.show()
```

```python
fig = px.histogram(df, x="damage", color="date_destruction")
fig.show()
```

```python
fig = px.histogram(df, x="damage", color="library_location")
fig.show()
```

```python
fig = px.histogram(df, x="gender", color="contribution")
fig.show()
```

```python
fig = px.histogram(df, x="gender", y="numbered_loss", histfunc='avg')
fig.show()
```
