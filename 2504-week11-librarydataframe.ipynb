```python
import requests
from bs4 import BeautifulSoup

# parser
import lxml
import pandas as pd
```

```python
site = requests.get('https://librarianswithpalestine.org/gaza-report-2024/')
source_code = site.content
libraries = BeautifulSoup(source_code, 'lxml')
```
